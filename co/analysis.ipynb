```python
import matplotlib.pyplot as plt 
import numpy as np 

%matplotlib widget

slab_co = np.loadtxt('slab_co/slab_co_energies.txt')
slab = np.loadtxt('../co2/slab/slab_energies.txt')
co_g = -14.937527
potential = slab_co[:,0]

fig = plt.figure()
ax = fig.add_subplot()
ax.plot(potential, slab_co[:,1] - slab[:,1] - co_g, marker='o', color='blue', fillstyle='none')
ax.set_ylabel(r"$\Delta E_\mathrm{elec}$ / eV")
ax.set_xlabel(r"$-\tilde \mu_\mathrm{e}$ / eV")
fig.savefig('electronic_be.png')
plt.show()
```
